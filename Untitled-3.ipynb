```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("iris-data.csv")
data.plot(kind='hexbin', x='sepal_length_cm', y='sepal_width_cm', gridsize=20)
```

```python
lengths = np.array(data['sepal_length_cm'])
lengths.sort()
widths = np.array(data['sepal_width_cm'])
widths.sort()
```

```python
X, Y = np.meshgrid(lengths, widths)
Z = np.sin(X) ** 10 + np.cos(10 + Y * X) * np.cos(X)
plt.contour(X,Y,Z, colors='green');
```

```python
sns.violinplot( x=data["iris_class"], y=data["sepal_length_cm"], linewidth=5)

sns.violinplot( x=data["iris_class"], y=data["sepal_length_cm"], width=0.3)
sns
```

```python
sns.violinplot( x=data["iris_class"], y=data["sepal_width_cm"], linewidth=5)

sns.violinplot( x=data["iris_class"], y=data["sepal_width_cm"], width=0.3)
sns
```
